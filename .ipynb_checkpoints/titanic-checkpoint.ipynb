```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_data=pd.read_csv("train.csv")
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_data[['Pclass', 'Survived']].groupby(['Pclass']).mean()
```

```python
train_data[['child','Survived']].groupby(['child']).mean()
```
